```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from tdc.single_pred import ADME
```

```python
dataset_names = ['HIA_Hou', 'BBB_Martins', 'CYP3A4_Substrate_CarbonMangels']
```

```python
dfs = []
for name in dataset_names:
    dfs.append(ADME(name=name).get_data())
    
```

```python
dfs[0].head()
```

```python
# any nans?
[df.isna().sum().sum() for df in dfs]
```

```python
# plot balance
plt.figure(figsize=(12, 4))
for i, df in enumerate(dfs):
    plt.subplot(1, len(dfs), i+1)
    df['Y'].hist()
    plt.title(dataset_names[i])
```
